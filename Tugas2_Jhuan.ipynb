```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_excel('C:/Users/ARANYA/Downloads/DATASET CYBERBULLYING INSTAGRAM - FINAL.xlsx')
df
```

```python
data = df[['Komentar']].sample(n = 3, replace = True).reset_index()
data = data[['Komentar']]
```

```python
data['lowercase'] = data['Komentar'].str.lower()
data
```

```python
data['punctual'] = data['lowercase'].str.replace('[^a-zA-Z]+',' ', regex=True)
data
```

```python
urlkamus = "https://raw.githubusercontent.com/jhuanss2/Lab_PML123/main/kamus_singkatan.csv"
kamus = pd.read_csv(urlkamus,sep=',')

kamus
```

```python
dict_kamus = kamus[['singkatan','asli']].to_dict('list')
dict_kamus
```

```python
dict_kamus = dict(zip(dict_kamus['singkatan'], dict_kamus['asli']))
dict_kamus
```

```python
norm = []

# Dengan perulangan, setiap kalimat akan dipecah menjadi kata2. 
# Kata2 yg terdapat pada dictionary langsung ditukar.
for i in data['punctual']:
    res = " ".join(dict_kamus.get(x, x) for x in str(i).split())
    norm.append(str(res))

# Memasukkan list kalimat normal kembali ke dalam dataframe
data['normalize'] = norm
data
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('indonesian'))
stop_words
```

```python
hapus = {'tidak','saya','tetapi','meskipun','mirip','dia','belum','kenapa','kalian','semua','ada','kurang','jadi'}
for x in hapus:
    stop_words.remove(x)
```

```python
swr = []

for i in data['normalize']:
    # Setiap kalimat akan dipecah menjadi kata2.
    tokens = word_tokenize(i)
    # Setiap kata akan dimasukkan ke list 'filtered' jika tidak terdapat pada stop_words.
    filtered = [word for word in tokens if word not in stop_words]
    # Kata2 pada list 'filtered' akan digabungkan kembali menjadi kalimat.
    swr.append(" ".join(filtered))

# Memasukkan list kalimat yg bersih kembali ke dalam dataframe
data['stopwords'] = swr
data
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Mendefenisikan stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Membuat list baru untuk menampung kalimat2 yang sudah di-stemming
stem = []

# Dari perulangan, stemmer mengganti setiap kata menjadi stem-nya.
for i in data['stopwords']:
    stem.append(stemmer.stem(str(i)))
    
# Memasukkan list kalimat yg sudah di-stem kembali ke dalam dataframe
data['stemmed'] = stem
data
```

```python
from sklearn.model_selection import train_test_split

data_training, data_testing = train_test_split(data, test_size=0.2, random_state=42, shuffle=True)

data_training = data_training.reset_index()
data_testing = data_testing.reset_index()

print('Jumlah data training: ', data_training.shape[0])
print('Jumlah data testing: ', data_testing.shape[0])
```

```python
MAX_NB_WORDS = 100000 #Maximum jumlah kata pada vocabulary yang akan dibuat
max_seq_len = 58 #Panjang kalimat maximum
```

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
```

```python
def tokenize_training(data_training):
    global tokenizer #Menggunakan variabel global agar 'tokenizer' bisa dipake di luar fungsi ini
    tokenizer = Tokenizer(num_words = MAX_NB_WORDS, char_level=False)
    tokenizer.fit_on_texts(data_training['stemmed'])
    word_index = tokenizer.word_index
    
    train_sequences = tokenizer.texts_to_sequences(data_training['stemmed'])
    word_seq_train = pad_sequences(train_sequences, maxlen = max_seq_len)

    return word_index, word_seq_train

word_index, word_seq_train = tokenize_training(data)
data['tokenize'] = word_seq_train.tolist()
data
```
